###  Data Preprocessing



```python
# Drop columns with excessive missing values (>70%)
threshold = 0.70
missing_fraction = train_df.isnull().mean()
drop_list = missing_fraction[missing_fraction > threshold].index.tolist()

print(f"Columns dropped due to high missingness (>70%): {len(drop_list)}")

# Apply the same reduction to both training and test data
X_train_reduced = X_train.drop(columns=drop_list)
X_test_reduced = X_test.drop(columns=drop_list)

print(f"Remaining feature count: {X_train_reduced.shape[1]}")
```

```python
# Create an aggregated feature from all ag_* histogram bins
ag_cols = [col for col in X_train_reduced.columns if col.startswith('ag_')]

X_train_reduced['ag_total'] = X_train_reduced[ag_cols].sum(axis=1)
X_test_reduced['ag_total'] = X_test_reduced[ag_cols].sum(axis=1)

print("Created feature 'ag_total'.")

# Visualize how the new feature differs by class
temp_df = X_train_reduced.copy()
temp_df['target'] = train_df['target']

plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='ag_total', data=temp_df)
plt.yscale('log')
plt.title('Distribution of ag_total by Class', fontsize=16)
plt.xlabel('Class (0 = No Failure, 1 = Failure)')
plt.ylabel('Total Histogram Count (log scale)')
plt.show()
```

```python
# Trainvalidation split and preprocessing pipeline
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler

# Extract the label column (map 'neg'/'pos' to 0/1)
target_col = None
for col in y_train.columns:
    if y_train[col].astype(str).str.contains('pos|neg').any():
        target_col = col
        break

y_clean = y_train[target_col].map({'neg': 0, 'pos': 1, '0': 0, '1': 1})

# Split into training and validation sets (stratified)
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_reduced,
    y_clean,
    test_size=0.2,
    random_state=42,
    stratify=y_clean
)

print("Data split complete.")
print(f"Train set: {X_train_split.shape}")
print(f"Validation set: {X_val_split.shape}")

# Preprocessing: median imputation + robust scaling
preprocessing_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', RobustScaler())
])

preprocessing_pipeline.fit(X_train_split)

# Apply preprocessing to all sets
X_train_clean = preprocessing_pipeline.transform(X_train_split)
X_val_clean   = preprocessing_pipeline.transform(X_val_split)
X_test_clean  = preprocessing_pipeline.transform(X_test_reduced)

# Convert back to DataFrames
X_train_clean_df = pd.DataFrame(X_train_clean, columns=X_train_reduced.columns)
X_val_clean_df   = pd.DataFrame(X_val_clean,   columns=X_train_reduced.columns)
X_test_clean_df  = pd.DataFrame(X_test_clean,  columns=X_train_reduced.columns)

# Final aligned targets
y_train_final = y_train_split.reset_index(drop=True)
y_val         = y_val_split.reset_index(drop=True)

print("Preprocessing complete (imputation and scaling applied).")
```

```python
# Handle class imbalance using SMOTE (applied only to the training set)
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_clean_df, y_train_final)

# Convert back to DataFrame/Series for consistency
X_train_resampled = pd.DataFrame(X_train_resampled, columns=X_train_clean_df.columns)
y_train_resampled = pd.Series(y_train_resampled, name='class')

print("SMOTE applied to training data.")
print(f"Resampled training shape: {X_train_resampled.shape}")
print("Class distribution after resampling:")
print(y_train_resampled.value_counts())
```

```python
# Feature selection based on correlation strength with the target

# Compute correlation of each feature with the class label
correlations = X_train_resampled.corrwith(y_train_resampled).abs()

# Keep features with correlation above the chosen threshold
corr_threshold = 0.05
selected_features = correlations[correlations > corr_threshold].index.tolist()

# Apply the selection to training and validation sets
X_train_final = X_train_resampled[selected_features]
y_train_final = y_train_resampled
X_val = X_val_clean_df[selected_features]

print("Correlation-based feature selection complete.")
print(f"Threshold: > {corr_threshold}")
print(f"Selected features: {len(selected_features)} of {X_train_resampled.shape[1]}")
print(f"Training shape:   {X_train_final.shape}")
print(f"Validation shape: {X_val.shape}")
```

### Hyper-parameter Tuning

```python
# Hyperparameter tuning for selected models with leakage-free SMOTE inside CV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, balanced_accuracy_score, confusion_matrix
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier

print("Tuning selected models...")

# Use clean (pre-SMOTE) data for tuning; SMOTE will be applied inside each CV fold
cols_to_use = X_train_final.columns
X_tune = X_train_clean_df[cols_to_use]
y_tune = y_train_split.reset_index(drop=True)

scorer = make_scorer(balanced_accuracy_score)
tuned_models = {}

def get_pipeline(model):
    """Wrap a classifier with SMOTE in a pipeline for leakage-free CV."""
    return ImbPipeline([
        ('smote', SMOTE(random_state=42)),
        ('clf', model)
    ])

# Model definitions and parameter grids
params_grid = {
    "Random Forest": (
        RandomForestClassifier(random_state=42, class_weight='balanced'),
        {
            'clf__n_estimators': [100, 200],
            'clf__max_depth': [10, 20, None],
        },
    ),
    "Extra Trees": (
        ExtraTreesClassifier(random_state=42, n_jobs=-1, class_weight='balanced'),
        {
            'clf__n_estimators': [100, 200],
            'clf__min_samples_split': [2, 5],
        },
    ),
    "XGBoost": (
        xgb.XGBClassifier(
            use_label_encoder=False,
            eval_metric='logloss',
            random_state=42,
            scale_pos_weight=59,
        ),
        {
            'clf__learning_rate': [0.05, 0.1],
            'clf__max_depth': [3, 5, 7],
        },
    ),
    "LightGBM": (
        lgb.LGBMClassifier(
            random_state=42,
            verbose=-1,
            scale_pos_weight=59,
        ),
        {
            'clf__learning_rate': [0.05, 0.1],
            'clf__num_leaves': [31, 50],
        },
    ),
    "Gradient Boost": (
        GradientBoostingClassifier(random_state=42),
        {
            'clf__learning_rate': [0.05, 0.1],
            'clf__max_depth': [3, 5],
        },
    ),
}

plt.figure(figsize=(15, 10))

for i, (name, (model, params)) in enumerate(params_grid.items()):
    print(f"Tuning {name}...")
    pipeline = get_pipeline(model)

    search = RandomizedSearchCV(
        pipeline,
        params,
        n_iter=5,
        scoring=scorer,
        cv=3,
        n_jobs=-1,
        random_state=42,
    )
    search.fit(X_tune, y_tune)

    best_pipe = search.best_estimator_
    tuned_models[name] = best_pipe

    # Evaluate on validation set
    X_val_tune = X_val[cols_to_use]
    y_pred = best_pipe.predict(X_val_tune)
    score = balanced_accuracy_score(y_val, y_pred)

    plt.subplot(2, 3, i + 1)
    sns.heatmap(
        confusion_matrix(y_val, y_pred),
        annot=True,
        fmt='d',
        cmap='Greens',
        cbar=False,
    )
    plt.title(f'Tuned {name}\nBal-Acc: {score:.4f}')
    plt.ylabel('True class')
    plt.xlabel('Predicted class')

plt.tight_layout()
plt.show()
```

### Combined Approaches

```python
# Build a weighted soft-voting ensemble of the tuned models
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import balanced_accuracy_score, confusion_matrix

print("Building weighted ensemble (Ensemble 5)...")

# Give higher weight to XGBoost and LightGBM
# Order in tuned_models: [Random Forest, Extra Trees, XGBoost, LightGBM, Gradient Boost]
weights = [1, 1, 2.5, 2.5, 1]

ensemble = VotingClassifier(
    estimators=[(name, model) for name, model in tuned_models.items()],
    voting='soft',
    weights=weights,
    n_jobs=-1
)

# Fit on the clean (pre-SMOTE) training data;
# each tuned model already applies SMOTE internally in its pipeline
ensemble.fit(X_tune, y_tune)

# Evaluate on validation data
X_eval = X_val
y_pred = ensemble.predict(X_eval)
score = balanced_accuracy_score(y_val, y_pred)
fn = confusion_matrix(y_val, y_pred)[1, 0]

print(f"Ensemble 5 balanced accuracy: {score:.5f}")
print(f"Ensemble 5 false negatives:   {fn}")

# Confusion matrix for the ensemble
plt.figure(figsize=(5, 4))
sns.heatmap(
    confusion_matrix(y_val, y_pred),
    annot=True,
    fmt='d',
    cmap='Purples',
    cbar=False
)
plt.title(f'Weighted Ensemble 5\nBal-Acc: {score:.4f}')
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()

# Compare ensemble performance with its components
comp_data = []
for name, model in tuned_models.items():
    pred = model.predict(X_eval)
    acc = balanced_accuracy_score(y_val, pred)
    fn_count = confusion_matrix(y_val, pred)[1, 0]
    comp_data.append({
        'Model': name,
        'Balanced Accuracy': acc,
        'False Negatives': fn_count
    })

comp_data.append({
    'Model': 'Ensemble 5 (Weighted)',
    'Balanced Accuracy': score,
    'False Negatives': fn
})

df_comp = pd.DataFrame(comp_data)

# Balanced accuracy comparison
plt.figure(figsize=(10, 6))
ax1 = sns.barplot(x='Model', y='Balanced Accuracy', data=df_comp, palette='viridis')
plt.ylim(0.5, 1.0)
plt.title('Balanced Accuracy: Components vs Ensemble 5', fontsize=14)
for p in ax1.patches:
    ax1.annotate(
        f'{p.get_height():.4f}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        fontsize=10
    )
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# False negatives comparison
plt.figure(figsize=(10, 6))
ax2 = sns.barplot(x='Model', y='False Negatives', data=df_comp, palette='magma')
plt.title('Missed Failures (False Negatives)', fontsize=14)
for p in ax2.patches:
    ax2.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center',
        va='bottom',
        fontsize=10
    )
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Super-Learner using OOF predictions to learn optimal ensemble weights
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, fbeta_score
from scipy.optimize import minimize
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Building Super-Learner using OOF predictions...")

# 1. Generate out-of-fold (OOF) probabilities for each tuned model
model_names = list(tuned_models.keys())
oof_preds_dict = {}

for name in model_names:
    print(f"Generating OOF predictions for {name}...")
    model = tuned_models[name]
    oof_probs = cross_val_predict(
        model,
        X_tune,
        y_tune,
        cv=3,
        method='predict_proba',
        n_jobs=-1
    )[:, 1]
    oof_preds_dict[name] = oof_probs

# Stack OOF predictions into a matrix
P_train_oof = np.column_stack([oof_preds_dict[name] for name in model_names])

# 2. Define objective: maximize F2-score on OOF data (equivalently minimize negative F2)
def safety_objective_oof(weights):
    weights = np.abs(weights)
    weights = weights / weights.sum()
    weighted_p = np.dot(P_train_oof, weights)
    final_pred = (weighted_p >= 0.5).astype(int)
    return -fbeta_score(y_tune, final_pred, beta=2)

# 3. Optimize weights subject to simple bounds
init_weights = np.ones(len(model_names)) / len(model_names)
bounds = [(0, 1)] * len(model_names)

opt_result = minimize(
    safety_objective_oof,
    init_weights,
    method='SLSQP',
    bounds=bounds
)
best_weights = opt_result.x / opt_result.x.sum()

print("\nOptimized ensemble weights (OOF-based):")
for name, w in zip(model_names, best_weights):
    print(f"{name:<15}: {w:.4f}")

# 4. Train final Super-Learner on the full clean training data
print("\nTraining Super-Learner on full training data...")
super_model = VotingClassifier(
    estimators=[(name, tuned_models[name]) for name in model_names],
    voting='soft',
    weights=best_weights,
    n_jobs=-1
)
super_model.fit(X_tune, y_tune)

# 5. Evaluate on the validation set (kept untouched during weight optimization)
X_eval = X_val[X_tune.columns]
y_pred_super = super_model.predict(X_eval)
super_score = balanced_accuracy_score(y_val, y_pred_super)
super_fn = confusion_matrix(y_val, y_pred_super)[1, 0]

print(f"\nSuper-Learner validation balanced accuracy: {super_score:.5f}")
print(f"Super-Learner false negatives: {super_fn}")

# Confusion matrix for the Super-Learner
plt.figure(figsize=(5, 4))
sns.heatmap(
    confusion_matrix(y_val, y_pred_super),
    annot=True,
    fmt='d',
    cmap='RdPu',
    cbar=False
)
plt.title(f'Super-Learner\nBal-Acc: {super_score:.4f}')
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

```python
# Final Star Model: blend ensemble and tuned models with threshold tuning
from sklearn.metrics import balanced_accuracy_score, confusion_matrix

# Validation data for evaluation
X_eval = X_val[X_tune.columns]

# Probabilities from ensemble components
p_ens   = ensemble.predict_proba(X_eval)[:, 1]
p_super = super_model.predict_proba(X_eval)[:, 1]
p_xgb   = tuned_models['XGBoost'].predict_proba(X_eval)[:, 1]
p_lgb   = tuned_models['LightGBM'].predict_proba(X_eval)[:, 1]

# Weighted blend of probabilities
# Ensembles receive higher weight; individual models act as correction
star_probs = (3 * p_ens + 3 * p_super + 1 * p_xgb + 1 * p_lgb) / 8

# Threshold search: prioritize low false negatives, then balanced accuracy
thresholds = np.arange(0.20, 0.60, 0.01)
best_threshold = 0.50
best_score = -1
best_fn = float('inf')

for t in thresholds:
    preds = (star_probs >= t).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_val, preds).ravel()
    acc = balanced_accuracy_score(y_val, preds)

    # Primary constraint: FN <= 10; secondary: maximize balanced accuracy
    if fn <= 10:
        if acc > best_score:
            best_score = acc
            best_threshold = t
            best_fn = fn
    elif best_score == -1 and fn < best_fn:
        # Fallback if no threshold satisfies FN <= 10
        best_fn = fn
        best_threshold = t
        best_score = acc

# Final performance at the chosen threshold
final_preds = (star_probs >= best_threshold).astype(int)
final_score = balanced_accuracy_score(y_val, final_preds)
conf_matrix = confusion_matrix(y_val, final_preds)
fn_final = conf_matrix[1, 0]
fp_final = conf_matrix[0, 1]

print("Star Model performance:")
print(f"Optimal threshold: {best_threshold:.2f}")
print(f"Balanced accuracy: {final_score:.5f}")
print(f"False negatives:   {fn_final}")
print(f"False positives:   {fp_final}")

# Confusion matrix plot
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title(f"Star Model (threshold = {best_threshold:.2f})")
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

### Prediction Performances and Overall View

```python
# Final model leaderboard on the validation set
import pandas as pd
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, recall_score

print("Generating final model leaderboard...")

# Validation data (use the same feature set as during tuning)
if 'X_val_tune' in globals():
    X_eval = X_val_tune
elif 'X_tune' in globals():
    X_eval = X_val[X_tune.columns]
else:
    X_eval = X_val

results_data = []

def add_record(name, model_type, preds):
    """Add a single model's metrics to the results list."""
    acc = balanced_accuracy_score(y_val, preds)
    cm = confusion_matrix(y_val, preds)

    if cm.shape == (2, 2):
        tn, fp, fn, tp = cm.ravel()
    else:
        # Fallback for degenerate cases
        tn = cm[0, 0] if cm.shape[0] > 0 else 0
        tp = cm[1, 1] if cm.shape[0] > 1 and cm.shape[1] > 1 else 0
        fn = 0
        fp = 0

    rec = recall_score(y_val, preds, zero_division=0)

    results_data.append({
        'Model Name': name,
        'Type': model_type,
        'Balanced Accuracy': acc,
        'False Negatives': fn,
        'False Positives': fp,
        'Recall (Safety)': rec,
    })

# Baseline models
if 'model_zoo_baseline' in globals():
    for name, model in model_zoo_baseline.items():
        try:
            preds = model.predict(X_eval)
        except Exception:
            preds = model.predict(X_val)
        add_record(name, 'Baseline', preds)

# Fine-tuned models
if 'tuned_models' in globals():
    for name, model in tuned_models.items():
        preds = model.predict(X_eval)
        add_record(name, 'FineTuned', preds)

# Ensemble 5
if 'ensemble' in globals():
    preds = ensemble.predict(X_eval)
    add_record('Ensemble5', 'Ensemble', preds)

# Super-Learner
if 'super_model' in globals():
    preds = super_model.predict(X_eval)
    add_record('SuperLearner', 'Super-Learner', preds)

# Star Model
if 'star_probs' in globals() and 'best_threshold' in globals():
    preds = (star_probs >= best_threshold).astype(int)
    add_record('StarModel', 'Final', preds)
elif 'final_preds' in globals():
    add_record('StarModel', 'Final', final_preds)

# Create and sort table
df_results = pd.DataFrame(results_data)
df_sorted = df_results.sort_values(by='Balanced Accuracy', ascending=False).reset_index(drop=True)

print("\nFinal model rankings:")
display(df_sorted)
```

```python
# --- Plot Top 10 Models by Balanced Accuracy (Simple Bar Chart) ---
import matplotlib.pyplot as plt
import seaborn as sns

# Take top 10 models from the sorted leaderboard
top10 = df_sorted.head(10)

plt.figure(figsize=(12, 6))

ax = sns.barplot(
    data=top10,
    x='Model Name',
    y='Balanced Accuracy',
    color='steelblue'
)

plt.title('Top 10 Models by Balanced Accuracy', fontsize=16)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Balanced Accuracy', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Annotate values on top of each bar
for bar in ax.patches:
    height = bar.get_height()
    ax.annotate(
        f'{height:.4f}',
        (bar.get_x() + bar.get_width() / 2, height),
        ha='center', va='bottom',
        fontsize=10, xytext=(0, 4),
        textcoords='offset points'
    )

plt.tight_layout()
plt.show()
```
